```python
import pandas as pd
import numpy as np
```

```python
url = 'https://api.covid19india.org/csv/latest/state_wise_daily.csv'
states = pd.read_csv(url, parse_dates= ['Date'],squeeze = True).sort_index()
states.head()
```

```python
pd = states[states['Status']== 'Confirmed']
```

```python
pd.head()
```

```python
pd1 = pd.drop('TT', axis = 1)
```

```python
pd1.head()
```

```python
pd1 = pd1.T
```

```python
pd1.head()
```

```python
pd1.drop('Status', axis = 0 ,inplace = True)
```

```python
pd1 = pd1.T
```

```python
pd1.head()
```

```python
pd1 = pd1.T
```

```python
pd1 = pd1.reset_index(inplace = False)
```

```python
pd1.head()
```

```python
new_header = pd1.iloc[0] 

pd1 = pd1[1:] 

pd1.columns = new_header
```

```python
pd1.head()
```

```python
pd1.index.name = 'name'
pd1.head()
```

```python
pd1 = pd1.rename(columns = {'Date':'name'})
```

```python
pd1.head()
```

```python
df = pd1.melt(id_vars=["name"], 
        var_name="Date", 
        value_name="Value")
```

```python
df.head(100)
```

```python
df = df.rename(columns = {'name':'state','Value':'positive'})
```

```python
df = df.rename(columns = {'positive':'cases'})
```

```python
df = df.rename(columns = {'Date':'date'})
```

```python
df
```

```python
df2 = df.sort_values(by = 'state')
df2.head()
```

```python
df2.head(1000)
```

```python
df2 = df2.reset_index(drop = True )
```

```python
df2.head(100)
```

```python
df2.to_csv('Covid-19')
```
